### Cuaderno de Clase: las Transformaciones de Herglotz

Las transformaciones de Herglotz corresponden a la versión general (con una velocidad que apunta en cualquier dirección) de las transformación de Lorentz-Einsteins que se escriben en la configuración estándar (con el eje x coincidiendo con la dirección de la velocidad relativa entre los dos sistemas). Las transformaciones vienen dadas por:

$$
\left(\begin{array}{c}T'\\x'\\y'\\z'\end{array}\right) = \left(\begin{array}{cccc}
\gamma & -\beta_x \gamma & -\beta_y \gamma & -\beta_z \gamma \\
-\beta_x \gamma & \frac{\beta_z^2 \gamma^2}{\gamma+1}+1 & \frac{\beta_x \beta_y \gamma^2}{\gamma+1} & \frac{\beta_z \beta_z \gamma^2}{\gamma+1} \\
-\beta_y \gamma & \frac{\beta_x \beta_y \gamma^2}{\gamma+1} & \frac{\beta_x^2 \gamma^2}{\gamma+1}+1 & \frac{\beta_y \beta_z \gamma^2}{\gamma+1} \\
-\beta_z \gamma & \frac{\beta_x x_z \gamma^2}{\gamma+1} & \frac{\beta_y \beta_z \gamma^2}{\gamma+1} & \frac{\beta_z^2 \gamma^2}{\gamma+1}+1
\end{array}\right)\left(\begin{array}{c}T\\x\\y\\z\end{array}\right)
$$

donde $T=ct$ es la variable temporal generalizada, $\beta_i = u_i/c$ es la componente cartesiana $i$-esima y $\gamma = 1 / \sqrt{1-\beta^2 }$ es el factor de Lorentz.

Escribamos una rutina para calcular las transformaciones de Herglotz:

```python
def herglotz_transformation(T, x, y, z, u_x, u_y, u_z, c=1):
    """
    Calculate Herglotz transformation for a 4-vector (T, x, y, z)
    with velocity components (u_x, u_y, u_z)
    
    Parameters:
    - T, x, y, z: coordinates in the original frame
    - u_x, u_y, u_z: velocity components
    - c: speed of light (default c=1 in natural units)
    
    Returns:
    - T', x', y', z': transformed coordinates
    """
    
    # Calculate beta components
    beta_x = u_x / c
    beta_y = u_y / c
    beta_z = u_z / c
    
    # Calculate beta squared
    beta_squared = beta_x**2 + beta_y**2 + beta_z**2
    
    # Calculate Lorentz factor
    gamma = 1 / np.sqrt(1 - beta_squared)
    
    # Build the Herglotz transformation matrix
    M = np.array([
        [gamma, -beta_x*gamma, -beta_y*gamma, -beta_z*gamma],
        [-beta_x*gamma, (beta_y**2 + beta_z**2)*gamma**2/(gamma+1) + 1, 
         beta_x*beta_y*gamma**2/(gamma+1), beta_x*beta_z*gamma**2/(gamma+1)],
        [-beta_y*gamma, beta_x*beta_y*gamma**2/(gamma+1), 
         (beta_x**2 + beta_z**2)*gamma**2/(gamma+1) + 1, beta_y*beta_z*gamma**2/(gamma+1)],
        [-beta_z*gamma, beta_x*beta_z*gamma**2/(gamma+1), 
         beta_y*beta_z*gamma**2/(gamma+1), (beta_x**2 + beta_y**2)*gamma**2/(gamma+1) + 1]
    ])
    
    # Apply transformation
    four_vector = np.array([T, x, y, z])
    transformed = M @ four_vector
    
    return transformed
```

Probamos la rutina:

```python
import numpy as np

# Test the Herglotz transformation with a simple example
# Let's use a point at (T=1, x=0, y=0, z=0) with velocity (0.5c, 0.3c, 0.2c)
T, x, y, z = 1, 0, 0, 0
u_x, u_y, u_z = 0.5, 0.3, 0.2

# Apply the transformation
T_prime, x_prime, y_prime, z_prime = herglotz_transformation(T, x, y, z, u_x, u_y, u_z)

print("Coordenadas originales:")
print(f"T = {T}, x = {x}, y = {y}, z = {z}")
print(f"\nVelocidad: u_x = {u_x}, u_y = {u_y}, u_z = {u_z}")
print(f"\nCoordenadas transformadas:")
print(f"T' = {T_prime:.4f}, x' = {x_prime:.4f}, y' = {y_prime:.4f}, z' = {z_prime:.4f}")
```
