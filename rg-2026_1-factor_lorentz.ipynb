### Cuaderno de Clase: el factor de Lorentz

```python
# Prompt: Dibuja en un gráfico semilogarítmico el factor de Lorentz en función de la velocidad 
import numpy as np

import matplotlib.pyplot as plt

# Definir el rango de velocidades (como fracción de c)
beta = np.linspace(0.001, 0.9999, 1000)  # v/c

# Calcular el factor de Lorentz: γ = 1/sqrt(1 - β²)
gamma = 1 / np.sqrt(1 - beta**2)

# Crear el gráfico log-log
plt.figure(figsize=(10, 6))
plt.semilogy(beta, gamma, linewidth=2)
plt.xlabel(r'$\beta = v/c$', fontsize=12)
plt.ylabel(r'$\gamma$ (Factor de Lorentz)', fontsize=12)
plt.title('Factor de Lorentz en escala log-log', fontsize=14)
plt.grid(True, which="both", ls="-", alpha=0.3)
plt.tight_layout()
plt.show()
```

**Nota:** Para velocidades cotidianas y astronómicas típicas, el factor de Lorentz es prácticamente 1, lo que significa que los efectos relativistas son despreciables. Solo cuando nos aproximamos significativamente a la velocidad de la luz (β > 0.1) los efectos relativistas se vuelven relevantes.
