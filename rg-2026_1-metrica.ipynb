### Cuaderno de Clase: la métrica

```python
import sympy as sp
from IPython.display import display
```

Primero definimos la forma de la matriz de rotación (transformación de isometría) en el espacio cuya métrica queremos encontrar (o verificar):

```python
teta = sp.symbols('theta')
R = sp.Matrix([[sp.cos(teta), sp.sin(teta)], [-sp.sin(teta), sp.cos(teta)]])
display(R)
```

Ahora postulamos una métrica cualquiera:

```python
g = sp.Matrix([[2, 1], [1, 3]]) 
display(g)
```

Ojo que la métrica siempre debe ser simétrica.

La prueba de si está si es la métrica del espacio donde la transformación de isometría es $R$ viene dada por la relación vista en clase:

$$
[g] = R^T [g] R
$$

```python
gtest = sp.simplify(R.T * g * R)

display(gtest)
```

Como vemos la nueva matriz no es idéntica a la matriz inicial lo que demuestra que no es la métrica. Probemos con la métrica ecuclidiana:

```python
g = sp.Matrix([[1, 0], [0, 1]])
gtest = sp.simplify(R.T * g * R)
display(gtest)
```

En este caso queda probado que efectivamente esta métrica no se modifica bajo la transformación de isometría y por lo tanto queda probado que está es la métrica de ese espacio. Pero hay una sorpresa: hay una libertad en la métrica que no conocíamos:

```python
a = sp.symbols('a')
g = sp.Matrix([[a, 0], [0, a]])
gtest = sp.simplify(R.T * g * R)
display(gtest)
```

En realidad la métrica euclidiana puede tener cualquier valor real en la diagonal. El valor de esta constante, sin embargo, no altera las propiedades geométrica y solo tiene que ver con la normalización de los vectores de base. Recordemos que:

$$
g_{ii} = ||e_i||^2
$$
así que la métrica con 1 en la diagonal simplemente es aquella en la que los vectores de base están normalizados.

## La métrica de Minkowski

Vamos a usando la transformación isométrica del espacio-tiempo (las transformaciones de Lorentz-Einsteins) a encontrar la métrica que satisface el espacio-tiempo. Comenzamos definiendo la TLEs en la configuración estándar:

```python
fi = sp.symbols('phi') # Esta es la rapidity
R = sp.Matrix([[sp.cosh(fi), -sp.sinh(fi)], [-sp.sinh(fi), sp.cosh(fi)]])
display(R)
```

Vamos a probar la métrica euclidiana como primera aproximación:

```python
g = sp.diag(*[1, 1]) # Sabemos que es diagonal
gtest = R.T * g * R
sp.simplify(gtest)
```

Es claro que el espacio-tiempo no tiene métrica euclidiana.

Vamos a probar una métrica diagonal de coeficientes arbitrario:

```python
mu, nu = sp.symbols('mu nu')
g = sp.diag(*[mu, nu]) # Sabemos que es diagonal
gtest = R.T * g * R
sp.simplify(gtest)
```

Como puede verse la única posibilidad para que la métrica resultante sea igual a la métrica original es si que cumple que $\mu+\nu=0$, es decir que $\mu=-\nu$:

```python
mu, nu = sp.symbols('mu nu')
g = sp.diag(*[mu, -mu]) # Sabemos que es diagonal
gtest = R.T * g * R
sp.simplify(gtest)
```

O bien:

```python
g = sp.diag(*[-mu, mu]) # Sabemos que es diagonal
gtest = R.T * g * R
sp.simplify(gtest)
```

De nuevo la escala aquí va a depender de la normalización de la base de modo que la métrica normal será:

```python
g = sp.diag(*[1, -1]) # Sabemos que es diagonal
gtest = R.T * g * R
sp.simplify(gtest)
```

Que es la métrica de Minkowski.

## Transformacione de Herglotz

Será muy interesante generalizar esto al caso de las transformaciones generales de Herglotz:

$$
\left(\begin{array}{c}T'\\x'\\y'\\z'\end{array}\right) = \left(\begin{array}{cccc}
\gamma & -\beta_x \gamma & -\beta_y \gamma & -\beta_z \gamma \\
-\beta_x \gamma & \frac{\beta_z^2 \gamma^2}{\gamma+1}+1 & \frac{\beta_x \beta_y \gamma^2}{\gamma+1} & \frac{\beta_z \beta_z \gamma^2}{\gamma+1} \\
-\beta_y \gamma & \frac{\beta_x \beta_y \gamma^2}{\gamma+1} & \frac{\beta_x^2 \gamma^2}{\gamma+1}+1 & \frac{\beta_y \beta_z \gamma^2}{\gamma+1} \\
-\beta_z \gamma & \frac{\beta_x x_z \gamma^2}{\gamma+1} & \frac{\beta_y \beta_z \gamma^2}{\gamma+1} & \frac{\beta_z^2 \gamma^2}{\gamma+1}+1
\end{array}\right)\left(\begin{array}{c}T\\x\\y\\z\end{array}\right)
$$


Use el procedimiento anterior para mostrar que la métrica para un espacio con esta transformación general tiene métrica $[g] = \mathrm{diag}(1,-1,-1,-1)$ o $[g] = \mathrm{diag}(-1,1,1,1)$
